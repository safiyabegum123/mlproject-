# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r"C:\Users\Shaik Safiya\Downloads\bike+sharing+dataset\hour.csv")
data.head()
```

```python
print(f'Dataset Shape: {data.shape}')
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['cnt'], bins=30, kde=True)
plt.title('Distribution of Bike Rentals')
plt.xlabel('Number of Rentals')
plt.ylabel('Frequency')
plt.show()
```

```python
categorical_features = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']

plt.figure(figsize=(15, 12))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data[feature])
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='hr', y='cnt', hue='weekday', palette='tab10')
plt.title('Bike Rentals vs. Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rentals')
plt.legend(title='Weekday')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='weathersit', y='cnt')
plt.title('Bike Rentals vs. Weather')
plt.xlabel('Weather Situation')
plt.ylabel('Number of Rentals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='temp', y='cnt', alpha=0.5)
plt.title('Bike Rentals vs. Temperature')
plt.xlabel('Normalized Temperature')
plt.ylabel('Number of Rentals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='season', y='cnt')
plt.title('Bike Rentals vs. Season')
plt.xlabel('Season')
plt.ylabel('Number of Rentals')
plt.show()
```

```python
g = sns.FacetGrid(data, col='season', row='weathersit', margin_titles=True)
g.map(sns.scatterplot, 'temp', 'cnt', alpha=0.5)
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Bike Rentals by Temperature, Season, and Weather')
plt.show()
```

```python
sns.pairplot(data[['cnt', 'temp', 'hum', 'windspeed']])
plt.show()
```
